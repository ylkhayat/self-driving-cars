## Sobel X / Y Threshold

```python
# X or Y sobel gradient
def abs_sobel_thresh(image, orient='x', sobel_kernel=3, thresh=(0, 255)):
    # TODO: Convert to grayscale using cv2.COLOR_RGB2GRAY as the conversion
    grayscale =  cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # TODO: Apply x or y gradient with the OpenCV Sobel() function
    # and take the absolute value
    if(orient == 'x'):
        sobel_value = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
    else:
        sobel_value = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
    sobel_value = np.absolute(sobel_value)
    # TODO: Rescale back to 8 bit integer
    abs_sobel = np.uint8(255*sobel_value/np.max(sobel_value))
    # TODO: Create a binary image of ones where threshold is met, zeros otherwise
    abs_sobel_output = np.zeros_like(abs_sobel)

    shape = abs_sobel_output.shape
    for i in range(0, shape[1]):
        for j in range(0, shape[0]):
            print(abs_sobel[i][j])
            if abs_sobel[i][j] <= thresh[1] and abs_sobel[i][j] >= thresh[0]:
                abs_sobel_output[i][j] = 1
            else:
                abs_sobel_output[i][j] = 0

    # Return the binary image
    return abs_sobel_output
```
